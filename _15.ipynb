```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Original dataset
data = [12, 15, 18, 20, 22, 22, 24, 25, 25, 25, 28, 30, 35, 40, 120]
df = pd.Series(data)

# Step 2: Summary statistics
mean = df.mean()
median = df.median()
mode = df.mode()[0]  # Take first mode if multiple

# Step 3: Quartiles and IQR
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Step 4: Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 5: Detect outliers
outliers = df[(df < lower_bound) | (df > upper_bound)]

# Step 6: Print summary results
print("Summary Statistics:")
print(f"Mean = {mean:.2f}")
print(f"Median = {median}")
print(f"Mode = {mode}")
print(f"Q1 = {Q1}")
print(f"Q3 = {Q3}")
print(f"IQR = {IQR}")
print(f"Lower Bound = {lower_bound}")
print(f"Upper Bound = {upper_bound}")
print(f"Outlier(s) = {outliers.values.tolist()}")

# Step 7: Visualization
plt.figure(figsize=(14, 6))

# Histogram with KDE
plt.subplot(1, 2, 1)
sns.histplot(df, bins=10, kde=True, color='skyblue', edgecolor='black')
plt.axvline(mean, color='red', linestyle='--', linewidth=1.5, label=f'Mean ≈ {mean:.2f}')
plt.axvline(median, color='green', linestyle='--', linewidth=1.5, label=f'Median = {median}')
plt.title("Histogram with KDE")
plt.xlabel("Monthly Sales (in $1000)")
plt.ylabel("Frequency")
plt.legend()

# Boxplot
plt.subplot(1, 2, 2)
sns.boxplot(x=df, color='lightgreen')
plt.title("Boxplot of Monthly Sales")
plt.xlabel("Monthly Sales (in $1000)")

plt.tight_layout()
plt.show()
```
