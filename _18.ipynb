data_dic = {'A':[12,15,45,71],

            'B':[18,26,50,80],

            'C':[32,48,78,93],

            'D':[29,35,63,81]}

Write the necessary code to prepare the following heatmap from above dictionary

unnamed.png

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_dic = {
    'A': [12, 15, 45, 71],
    'B': [18, 26, 50, 80],
    'C': [32, 48, 78, 93],
    'D': [29, 35, 63, 81]
}
df = pd.DataFrame(data_dic)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df, annot=True, cmap='rocket')
plt.tight_layout()
plt.show()
```
