# Question

Draw the line chart using “x” as the output given here

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(11)
y1 = x + 2
y2 = x + 3
print(y1, y2)
```

```python
plt.plot(x, y1, label="x+2")
plt.plot(x, y2, label="x+3")
plt.legend(loc="lower right")
plt.show()
```
