<p><h3>Write a Program to take a user score and find a student's grade using Python code. Following is the grade information.</h3></p>


image.png

```python
# Enter marks/scores of a student to find his grade.
marks = int(input(""))

if(marks > 100 or marks < 0):
  print("Invalid marks. Please enter a number between 0 and 100.")
elif(marks <= 100 and marks >= 90):
  print("A")
elif(marks <= 89 and marks >= 80):
  print("B")
elif(marks <= 79 and marks >= 70):
  print("C")
elif(marks <= 69 and marks >= 60):
  print("D")
elif(marks <= 59 and marks >= 50):
  print("E")
elif(marks <= 49 and marks >= 40):
  print("E-")
else:
  print("F or Fail")
```
