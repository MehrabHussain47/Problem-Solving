```python
import pandas as pd
import numpy as np
```

```python
data = {

    'CustomerID': [101, 102, 103, 104, 105, 106, 107],
    'Name': [' alice ', 'BOB', 'Charlie', None, 'Eve', '  frank', 'Grace'],
    'Age': ['25', 'thirty', '35', '40', None, '29', '27'],
    'City': ['New York', 'los angeles ', 'CHICAGO', 'New York', 'Chicago', None, 'new york'],
    'SignupDate': ['2022-01-01', '2022-02-15', None, '2022-03-01', 'invalid', '2022-03-15', '2022-03-20'],

}
```

```python
# Load the dictionary into a pandas DataFrame.
df = pd.DataFrame(data)
df
```

```python
# Strip whitespaces from Name and City.
df['Name'] = df['Name'].str.strip()
df['City'] = df['City'].str.strip()
df
```

```python
# Convert Name to title case.
df['Name'] = df['Name'].str.title()
df
```

```python
# Convert City to lowercase.
df['City'] = df['City'].str.lower()
df
```

```python
# Convert Age to numeric, setting errors to NaN.
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df
```

```python
# Convert SignupDate to datetime, coercing errors.
df['SignupDate'] = pd.to_datetime(df['SignupDate'], errors='coerce')
df
```

```python
# Drop rows where Name or City is missing.
df['City'] = df['City'].apply(
    lambda x: str(x).strip() if pd.notnull(x) and str(x).strip().lower() != 'none' else np.nan
)
df = df.dropna(subset=['Name', 'City'])
df
```

```python
# Fill missing Age with the median value.
df['Age'] = df['Age'].astype(float)
median_age = df['Age'].median()
df['Age'] = df['Age'].fillna(median_age)
df
```

```python
# Show only customers from new york with Age < 30 and valid signup dates (non-NaT).
filtered_df = df[
    (df['City'] == 'new york') &
    (df['Age'] < 30) &
    (df['SignupDate'].notna())
]
filtered_df
```

```python
# Display only CustomerID, Name, and Age columns.
result = filtered_df[['CustomerID', 'Name', 'Age']]
print(result)
```
